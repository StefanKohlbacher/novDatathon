```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
import json

randomState = 991
np.random.seed(randomState)
```

```python
salesTrain = pd.read_csv('./data/data_files/sales_train_splitted.csv', index_col=0)
salesTrain.head()
```

### Random split by regions

```python
nrFolds = 5
folds = {}
kfold = KFold(n_splits=nrFolds, shuffle=True, random_state=randomState)
for i, indices in enumerate(kfold.split(np.arange(151))): 
    trainIndex, testIndex = indices
    folds[i] = {
        'training': trainIndex.tolist(), 
        'test': testIndex.tolist()
    }
    
folds
```

```python
with open('./data/data_files/train_validation_split.json', 'w') as f: 
    json.dump(folds, f)
```
